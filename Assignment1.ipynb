```python
# 1 calculate the total bill for two items
Item1=int(input("Enter the price of first item:"))
Item1_quantity=int(input("Enter the first quantity:"))
Item2=int(input("Enter the price of second item:"))
Item2_quantity=int(input("Enter the quantity of second item:"))
Item1_price=Item1*Item1_quantity
Item2_price=Item2*Item2_quantity
Total=Item1_price+Item2_price
print("The total amount is:",Total)
```

```python
# 2 convert temperature from Celsius to Fahrenheit
C=int(input("Enter the temperature in celsius:"))
F=(C*9/5)+32
print("THe temperature in fehrenhiet:",F)
```

```python
# 3 calculate the total salary including HRA (20%) and DA (10%)
basic=int(input("Enter the basic salary:"))
HRA=0.2*basic
DA=0.1*basic
total_salary=basic+HRA+DA
print("The net salary:",total_salary)
```

```python
# 4 calculate simple interest using the formula SI = (P × R × T) / 100.
Principle=int(input("Enter the principle:"))
Rate_of_Intrest=int(input("Enter the intrest:"))
Time=int(input("Enter the time(In years):"))
Simple_Intrest=(Principle*Rate_of_Intrest*Time)/100
print("Simple Intrest:",Simple_Intrest)
```

```python
# 5 convert total minutes into hours and minutes
Total_min=int(input("The the total minutes:"))
Hour=Total_min//60
Min=Total_min%60
print(Hour,"hours and",Min,"minutes")
```

```python
# 6 calculate the Body Mass Index (BMI) using the formula: BMI weight(kg) / (height(m) × height(m))
weight=int(input("Enter the weight(kg):"))
height=float(input("Enter the height(m):"))
BMI=weight/(height*height)
print("BMI:",BMI)
```

```python
# 7 Write a program that takes the original price and prints the discounted price.
Original_Price=int(input("Enter the price:"))
Discount=0.1*Original_Price
D=Original_Price-Discount
print("Discounted Price:",D)
```

```python
# 8  area of an irregular triangular plot
a=float(input("Enter side a:"))
b=float(input("Enter side b:"))
c=float(input("Enter the side c:"))
s=(a+b+c)/2
area=(s*(s-a)*(s-b)*(s-c))**0.5
print("Area of triangle:",area)
```

```python
# 9 calculate the total cost to paint a rectangular wall
length=float(input("Enter the wall length:"))
height=float(input("Enter the wall height:"))
wall_area=length*height
paint_rate=25
window_lenght=1.5
window_height=1.2
windowarea=window_lenght*window_height
total_windowarea=2*windowarea
paint=wall_area-total_windowarea
total_cost=paint*paint_rate
print("Total area to be painted:",paint)
print("Total painting cost:₹",total_cost)
```
