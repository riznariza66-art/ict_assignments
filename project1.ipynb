1. Smart Library Circulation **System**

```python
from datetime import datetime

# Business rules
MIN_BALANCE = 1000
DAILY_LIMIT = 5000
accounts = {}

class Account:
    def __init__(self, name, acc_no):
        self.name = name
        self.acc_no = acc_no
        self.balance = 0
        self.txns = []
        self.daily_withdrawn = 0

    def deposit(self, amt):
        if amt <= 0: return print("Invalid amount")
        self.balance += amt
        self.txns.append(f"{datetime.now()} - Deposited: {amt}, Balance: {self.balance}")
        print(f"Deposited {amt}, Balance: {self.balance}")

    def withdraw(self, amt):
        if amt <= 0: return print("Invalid amount")
        if self.balance - amt < MIN_BALANCE: return print("Cannot withdraw, min balance required")
        if self.daily_withdrawn + amt > DAILY_LIMIT: return print("Daily withdrawal limit exceeded")
        self.balance -= amt
        self.daily_withdrawn += amt
        self.txns.append(f"{datetime.now()} - Withdrawn: {amt}, Balance: {self.balance}")
        print(f"Withdrawn {amt}, Balance: {self.balance}")

    def transfer(self, target, amt):
        if amt <= 0 or self.balance - amt < MIN_BALANCE: return print("Transfer failed")
        self.balance -= amt
        target.balance += amt
        self.txns.append(f"{datetime.now()} - Transferred {amt} to {target.acc_no}")
        target.txns.append(f"{datetime.now()} - Received {amt} from {self.acc_no}")
        print(f"Transferred {amt} to {target.acc_no}")

    def mini_statement(self):
        print(f"\n--- Mini Statement ({self.acc_no}) ---")
        for t in self.txns[-5:]: print(t)
        print(f"Balance: {self.balance}\n")

# Helper to get account
def get_acc(prompt="Enter account number: "):
    acc_no = input(prompt)
    acc = accounts.get(acc_no)
    if not acc: print("Account not found!")
    return acc

# Menu
def menu():
    while True:
        print("\n1.Create 2.Deposit 3.Withdraw 4.Transfer 5.Statement 6.Exit")
        choice = input("Choice: ")
        if choice == '1':
            name = input("Name: ")
            acc_no = input("Account No: ")
            if acc_no in accounts: print("Account exists")
            else: accounts[acc_no] = Account(name, acc_no)
        elif choice == '2':
            acc = get_acc(); acc and acc.deposit(float(input("Amount: ")))
        elif choice == '3':
            acc = get_acc(); acc and acc.withdraw(float(input("Amount: ")))
        elif choice == '4':
            acc1 = get_acc("Your account: "); acc2 = get_acc("Target account: ")
            if acc1 and acc2: acc1.transfer(acc2, float(input("Amount: ")))
        elif choice == '5':
            acc = get_acc(); acc and acc.mini_statement()
        elif choice == '6': break
        else: print("Invalid choice")

menu()
```
