```python
# Write a Python function to sum all the numbers in a list.
def sum(numbers):
    total = 0
    for x in numbers:
        total += x
    return total
numbers=[8,2,3,0,7]
print(sum(numbers))
```

```python
"""Python program to insert a number to any position in a list (Get the number
to be inserted and position from the user)"""
l1=['Apple',"Banana",'Orange']
n=input("Enter the fruit name:")
e=int(input("Enter the position to add:"))
l1.insert(e,n)
print(l1)
```
